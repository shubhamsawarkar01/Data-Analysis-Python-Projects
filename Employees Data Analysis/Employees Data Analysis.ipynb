```python
ax = sns.countplot(x='Gender',data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
salary_gen = df.groupby(['Gender'],as_index =False)['Salary'].sum().sort_values(by = 'Salary', ascending = False)
sns.barplot(x='Gender',y='Salary',data=salary_gen)
```

### From above plot we can see males and females are getting almost similar salaries.

```python
salary_gen = df.groupby(['Experience (Years)'],as_index =False)['Salary'].sum().sort_values(by = 'Salary', ascending = True)
sns.scatterplot(x='Experience (Years)',y='Salary',data=salary_gen)
```

### There is a noticeable positive correlation between years of experience and salary. As the number of years of experience increases, the salary tends to increase as well.

```python
salary_gen = df.groupby(['Position'],as_index =False)['Salary'].sum().sort_values(by = 'Salary', ascending = False)
sns.set(rc= {'figure.figsize' : (25,8)})
sns.barplot(x='Position',y='Salary',data=salary_gen)
```

### The graph suggests a trend where managerial and specialized technical roles command higher salaries, while more general or entry-level positions tend to have lower compensation.

```python
salary_gen = df.groupby(['Position'],as_index =False)['Experience (Years)'].sum().sort_values(by = 'Experience (Years)', ascending = False)
sns.set(rc= {'figure.figsize' : (25,8)})
sns.barplot(x='Position',y='Experience (Years)',data=salary_gen)
```

### Database Administrator: Shows the highest experience level, suggesting that this role typically requires extensive knowledge and expertise.

# Conclusion : This analysis highlights the importance of both experience and specialization in determining salary levels within the IT industry.
